# K-Means Clustering

## Importing the libraries

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
path = r'D:\Projects\Software_Engineering\Artificial_Intelligence\Huawei_Internship_16_Sep/Datasets'

dataset = pd.read_csv(os.path.join(path, 'Mall_Customers.csv'))
X = dataset.iloc[:, [3, 4]].values
```

## Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 10), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Training the K-Means model on the dataset

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 7, init = 'k-means++', random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```
