**Task_2**
ake 5 passwords from user and save them let user enter a password if it's one of the 5 print valid if not print invalid

```python
pass_list = []

for i in range(5):
  password = input(f'pls, enter password #{i+1}: ')
  pass_list.append(password)

entered_password = input('pls, enter password: ')

if entered_password in pass_list:
  print('valid')
else:
  print('invalid')
```
