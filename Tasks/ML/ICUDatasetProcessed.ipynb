```python
import pandas as pd
import os

path = r'C:\Users\Shimaa\Desktop\Huawei_Internship_16_Sep\Datasets\ICUDatasetProcessed'

file_list = os.listdir(path)

file_list
```

```python
df1 = pd.read_csv(os.path.join(path, file_list[0]))
df1.head()
```

```python
df2 = pd.read_csv(os.path.join(path, file_list[1]))
df2.head()
```

```python
df3 = pd.read_csv(os.path.join(path, file_list[2]))
df3.head()
```

```python
print(f'df1.shape: {df1.shape}')
print(f'df2.shape: {df2.shape}')
print(f'df3.shape: {df3.shape}')
```

```python
# Check if all DataFrames have the same columns
if (df1.columns.equals(df2.columns)) and (df2.columns.equals(df3.columns)):
    print("All three DataFrames have identical columns.")
else:
    print("The DataFrames have different columns.")
```

- axis=0 means that you are concatenating the DataFrames vertically (i.e., stacking them on top of each other). This combines the rows of the DataFrames, adding the rows from df2 below the rows from df1.
- we use -> ignore_index=True

- axis=1 would mean that you are concatenating the DataFrames horizontally (i.e., side by side). This combines the columns of the DataFrames, adding the columns from df2 next to the columns from df1.
- we use -> ignore_index=False

```python
merged_df = pd.concat([df1, df2, df3], axis=0, ignore_index=True)
merged_df.shape
```

```python
merged_df.info()
```

```python
num_df = merged_df.select_dtypes(include=['number'])
nan_df = merged_df.select_dtypes(exclude=['number'])

print(f'len(num_df.columns): {len(num_df.columns)}\nnum_df.columns\n{num_df.columns}')
print(f'len(nan_df.columns): {len(nan_df.columns)}\nnan_df.columns\n{nan_df.columns}')
```

```python
from sklearn.preprocessing import LabelEncoder

for col in nan_df.columns:
    nan_df[col] = nan_df[col].astype(str) # Convert all values in the column to string to ensure uniformity
    nan_df[col] = LabelEncoder().fit_transform(nan_df[col])

merged_df = pd.concat([num_df, nan_df], axis=1, ignore_index=False)

merged_df
```

```python
print(f'len(merged_df.columns): {len(merged_df.columns)}')

non_feats_col = []
for col in merged_df.columns:
    if len(merged_df[col].unique()) == 1:
        non_feats_col.append(col)

merged_df = merged_df.drop(labels=non_feats_col, axis=1, inplace=False)

print(f'len(non_feats_col): {len(non_feats_col)}')
print(f'len(merged_df.columns): {len(merged_df.columns)}')
```

```python
from sklearn.preprocessing import MinMaxScaler

X = merged_df.iloc[:, :-1].values
y = merged_df.iloc[:, -1].values

X = MinMaxScaler().fit_transform(X)

print(f'X.shape: {X.shape}')
print(f'y.shape: {y.shape}')
```

```python
from sklearn.feature_selection import SelectKBest, chi2

X = SelectKBest(chi2, k=30).fit_transform(X, y)
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

model = LogisticRegression()
model_result = model.fit(X_train, y_train)

y_pred = model_result.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)

print(f'confusion_matrix\n{cm}\n')
print(f'accuracy_score: {acc}')
```
